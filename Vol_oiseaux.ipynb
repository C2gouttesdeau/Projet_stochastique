## I - Marche aléatoire (oiseaux aveugles) : a = 0, η = 1

```python
#### Constante ####
l=10 #longueur de la boite
v0 = 0.03 #Vitesse
dt = 1 #Pas de temps
```

Les oiseaux aveugles ne voit pas leur voisins (a=0) donc leur trajectoire ne doivent pas dépendre de ceux qui les entours.
Si η = 1 alors leur mouvement est complétement aléatoire, si η = 0 alors les oiseaux ne vont que dans un seul direction sans jamais en changer ! 

### Calcul de la solution et illustration

```python
##### To change ######
a=0                  #echelle de longueur
N=10                #nombre de particules              
Nt=10             #Nombre d'itérations (pas de temps)
eta = 1              #eta : paramètre d'aléatoire

Save = True         #True pour sauvegarder l'animation
Trajectoires = True #True pour afficher les trajectoires (gérer pour N=1)
name = "a="+str(a)+"_N="+str(N)+"_eta="+str(eta)+"_v0="+str(v0)+"_Nt="+str(Nt)  #Nom de la simulation
######################
```

```python
###### Calcul de la solution spacio-temporelle #######
x_sol , y_sol, theta_sol = Solveur(N,l,a,v0,dt,eta,Nt) #Calcul des solutions
```

```python
name+'.gif'
```

```python
###### Mise en place de l'annimation #######
AnimationGif(x_sol,y_sol,N,l,Nt,100,name,Trajectoires,Save)
plt.close()
display(Image(name+'.gif'))
```

```python
###### Calcul de Phi et affichage #######
phi,Var_phi = Calc_Phi(N,theta_sol)
t = np.arange(0,Nt)
plt.figure(2)
plt.plot(t,phi,label="Phi");plt.plot(t,Var_phi,label="Variance de Phi")
plt.legend();plt.xlim(0);plt.ylim(0);plt.xlabel("Temps");plt.ylabel("Phi")
plt.figure(2)
plt.show()
```

1. Vérifiez qu’en prenant a = 0 (pas d’effet collectif) et η = 1, vous retrouvez une marche aléatoire
(mouvement brownien appelé aussi processus de Wiener). Cela ne convient peut-être pas au cas
du vol d’étourneaux, mais cela montre que le modèle de Vicsek contient le modèle brownien
d’agitation thermique (cas des micro-particules en suspension dans un liquide).

## II - Vols avec effets collectifs parfaits : η = 0

```python
##### To change ######
Nt = 100             #Nombre d'itérations (pas de temps)
######################

eta = 0              #eta : paramètre d'aléatoire
a=1                  #echelle de longueur
Save = True         #True pour sauvegarder l'animation
Trajectoires = False #True pour afficher les trajectoires (gérer pour N=1)
######################
```

```python
for i in np.linspace(10,200,4).astype(int):
    x_sol , y_sol, theta_sol = Solveur(N,l,a,v0,dt,eta,Nt) #Calcul des solutions
    name = "a="+str(a)+"_N="+str(i)+"_eta="+str(eta)+"_v0="+str(v0)+"_Nt="+str(Nt)  #Nom de la simulation
    AnimationGif(x_sol,y_sol,N,l,Nt,100,name,Trajectoires,Save)
    plt.close()
    display(Image(name+'.gif'))
```
